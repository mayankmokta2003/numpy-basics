```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr1 = np.array([[1,2,3],[4,5,6]])
arr2 = np.random.random((2,3))
arr3 = np.zeros((2,2))
```

```python
np.save("arr1.npy", arr1)
np.save("arr2.npy", arr2)
np.save("arr3.npy", arr3)
```

```python
load_arr1 = np.load("arr1.npy")
load_arr2 = np.load("arr2.npy")
load_arr3 = np.load("arr3.npy")
print("arr1: ", load_arr1)
print("arr2: ", load_arr2)
print("arr3: ", load_arr3)
```

## saving an image

```python
# suppose there is a file save we need to load in and show it through metplotlib
# here we presented a whole picture into numbers or matrices and we can read it and through pyplot we display it
try:
    load_img = np.load("File.npy")

    # Display
    plt.figure(figsize=(10,5))
    plt.subplot(121)
    plt.imshow(load_img)
    plt.title("Image logo")
    plt.grid(False)

    # Now to convert this image to darkmode
    dark_image = 1 - load_img
    plt.subplot(122)
    plt.title("Dark image")
    plt.grid(False)
    plt.imshow(dark_image)
    
except FileNotFoundError:
    print("File not found")
```
