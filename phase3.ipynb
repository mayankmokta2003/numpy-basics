```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Structure: [restraunt_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000], #restraunt 1
    [2, 120000, 140000, 160000, 190000], #restraunt 2
    [3, 200000, 230000, 260000, 300000], #restraunt 3
    [4, 180000, 210000, 240000, 270000], #restraunt 4
    [5, 160000, 185000, 205000, 230000]  #restraunt 5
])

print("Shape of data is: ", sales_data.shape)
print("Size of data is: ", sales_data.size)
print("Data of firse three restraunt: ", sales_data[0:3])
```

```python
# Total sales per year
total_sales = np.sum(sales_data[0:5,1:5], axis=0)
print("Yearly total: ",total_sales)
total = np.sum(total_sales)
print("Total sales: ", total)


# Minimum sales per restraunt
min_sales = np.min(sales_data[0:5,1:5], axis=1)
print("Minimum sales per restraunt: ", min_sales)


# Maximum sales per year
max_sales = np.max(sales_data[0:5, 1:], axis=0)
print("Maximum sales per year: ", max_sales)

# Average sales per restraunt
ave_sales = np.average(sales_data[:, 1:], axis=1)
print("Average sales per restraunt", ave_sales)
```

```python
## Cumelative sales of restraunt and create a graph of it 
cum_sales = np.cumsum(sales_data[:, 1:], axis=1)
print("Cumelative sales of restraunt: ", cum_sales)

plt.figure(figsize=(12,6))
plt.plot(np.mean(cum_sales, axis=0))
plt.title("Average cumelative sales across restraunts")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1,2,3,4])
vector2 = np.array([5,6,7,8])

add = np.add(vector1, vector2)
mul = np.multiply(vector1, vector2)
dot = np.dot(vector1, vector2)
angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))

print("Adding two vectors: ",add)
print("multiplying two vectors: ",mul)
print("Dot product of two vectors: ",dot)
print("Angle between vectors is: ", angle)
```

```python
## Vectorised operations

rest_types = np.array(["chinese", "indian", "burger", "pizza"])
vectorized_firse_upper = np.vectorize(str.capitalize)
vectorized_upper = np.vectorize(str.upper)
print("Vectorized first letter uppercase: ", vectorized_firse_upper(rest_types))
print("Vectorized string uppercase: ", vectorized_upper(rest_types))
```
