```python
import numpy as np
```

### Creating array from list

```python
arr_1d = np.array([1,2,3])
print("the 1d array is: ",arr_1d)

arr_2d = np.array([[1,2,3], [4,5,6]])
print("the 2d array is: ", arr_2d)
print(type(arr_1d))

```

```python
zeroArr = np.zeros((3,4))
print("Zeros Array: \n", zeroArr)

onesArr = np.ones((4,3))
print("\nOnes Array: \n", onesArr)

fullArr = np.full((3,4), 10)
print("\nFull value 10 Array: \n", fullArr)

randomArr = np.random.random((3,4))
print("\nRandom values Array: \n", randomArr)

sequence = np.arange(0,11,2)
print("\nSequence values Array: \n", sequence)

```

### VECTOR, MATRIX, TENSOR

```python
vector = np.array([1,2,3])
print("Vertor: \n", vector, vector.shape)

matrix = np.array([[1,2,3], 
                   [4,5,6]])
print("Matrix: \n", matrix, matrix.shape)

tensor = np.array([[[1,2], [3,4]],
                   [[5,6],[7,8]]])
print("Tensor: \n", tensor, tensor.shape)
```

### ARRAY PROPERTIES

```python
arr = np.array([[1,2,3],
                [4,5,6]])

print("Shape: ", arr.shape)
print("Dimension: ", arr.ndim)
print("Size: ", arr.size)
print("DataType: : ", arr.dtype)
```

### Array Reshaping

```python
arr = np.arange(12)
print("Original Array: \n", arr)

reshaped = arr.reshape((3,4))
print("Reshaped Array: \n", reshaped)

flattened = reshaped.flatten()
print("Flattened Array: \n", flattened)

# returns view instead of copy
raveled = reshaped.ravel()
print("Raveled Array: \n", raveled)

transpose = reshaped.T
print("Transpose Array: \n", transpose)
```
